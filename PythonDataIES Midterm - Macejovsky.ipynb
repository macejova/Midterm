```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
import matplotlib
```

```python
import pandas as pd
```

```python
''' Findings links to the individual books '''

response = requests.get('https://en.wikipedia.org/wiki/20th_Century%27s_Greatest_Hits:_100_English-Language_Books_of_Fiction')
soup = BeautifulSoup(response.text,'lxml')
div = soup.find("ol")  

import re
parenth=re.findall('\"[^"]+\"', str(div)) 
links1=parenth[0:len(parenth):2]  #geting partial links to the books inside of " "
links=[]
for x in links1:
    links.append(x.replace('"', ""))   # getting rid of " "

wiki='https://en.wikipedia.org'
wiki_links=[wiki+links[x] for x in range(len(links)-1)]  # final links
 
```

```python
''' Getting data from tables: '''

name_of_attributes=[]
attributes=[]
for link in wiki_links:
    response2=requests.get(link)
    soup2 = BeautifulSoup(response2.text)
    table=soup2.find("table",{'class':'infobox vcard'})
    try:
        name_of_attributes.append([th.text for th in table.find_all('th')])  # for example "Author"
    except:
        name_of_attributes.append("nothing")
    try:
        attributes.append([td.text for td in table.find_all('td')])   # for example "George Orwell"
    except:
        attributes.append("nothing")    
```

```python
 
titles=parenth[1:len(parenth):2]   
titles
df=[]
for title in range(len(titles)-1):
    df.append({titles[title]:attributes[title]})
df
pd.DataFrame(df)
```
